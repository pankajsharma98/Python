```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-5,5,1000)
y=abs(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
y=x**2
plt.plot(x,y)
plt.show()
```

```python
y=x**3
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,'r:',label='Line')
plt.show()
```

```python
plt.plot(x,x,'r:',label='Line')
#plt.show(x,y,'g--',lw=3)
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(10,10))
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.2,0.2,0.2])
ax3=fig.add_axes([0.6,0.2,0.2,0.2])
ax4=fig.add_axes([0.2,0.6,0.2,0.2])
ax5=fig.add_axes([0.6,0.6,0.2,0.2])
x=np.linspace(-1,1,1000)
ax2.plot(x,x,'r:',label="Line")
ax2.legend()
log=np.log(x)
ax3.plot(x,log,'b:',label="Log")
ax3.legend()
mod=abs(x)
ax4.plot(x,mod,'g:',label="Mod")
ax4.legend()
exp=np.exp(x)
ax5.plot(x,exp,'b:',label="exp")
ax5.legend()
plt.savefig("plots.png")
plt.show
```

```python
fig=plt.figure(figsize=(7,7))
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.1,0.57,0.4,0.4])
x=np.linspace(-1,1,1000)
ax1.plot(x,x,'r:',label="Line")
ax1.legend()
y=x**2
ax2.plot(y,'b:',label="$x^2$")
z=x**3
ax2.plot(z,'g:',label="$x^3$")
ax2.legend()
plt.title("xyz",fontsize=10)
plt.show()
```
