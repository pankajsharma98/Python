```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.arange(1,101)
y=np.random.normal(5.5,.8,100)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
w=np.random.normal(40,70,100)
h=w*np.random.normal(5.5,.8,100)
age=np.random.randint(18,50,100)
```

```python
plt.scatter(w,h,c=age)
plt.show()
```

```python
x=np.array([3,4,5,6,7,8,9,5,7,8])
y=np.array([6,5,7,4,8,3,7,6,9,6])
#mean=np.mean(x)
#print(mean)
#plt.plot([6.2,6.2],[0,9],c="r",lw=2)
plt.scatter(x,y)
plt.plot([x.mean(),x.mean()],[0,9],c='r',lw=2)
#plt.plot([1,x.mean()],[5,5],'r:',lw=2)
for index in range(len(x)):
    plt.plot([x[index],x.mean()],[y[index]]*2,'r--',lw=2)
plt.show()
```
