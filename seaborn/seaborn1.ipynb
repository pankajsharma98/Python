```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
x=[1,2,3,4,5,6]
y=[4,5,6,2,3,9]
```

```python
plt.plot(x,y)
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.scatter(x,y)
plt.setp(ax.xaxis.get_majorticklabels(),rotation=90)
ax.set_xticks([1,2,3,4,5])
ax.set_yticks(range(0,10))
ax.set_xticklabels(["India","Pakisthan","SouthAfrica","Australia","nepal"])
plt.show()
```

```python
import pandas as pd
```

```python
df=pd.read_csv("seaborn-data/tips.csv")
```

```python
df.head()
```

```python
fig=plt.figure(figsize=(5,3))
ax=plt.subplot(111)
sns.distplot(df['total_bill'],kde=False,bins=30,color="green",ax=ax)
ax.set_title("Distribution Plot",fontsize=20)
#plt.grid()
plt.show()
```

```python
sns.pairplot(df,hue='sex')
```
